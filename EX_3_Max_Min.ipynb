```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return 100+x**2+20*x

def derivative(x):
    return 2*x+20
    
def get_minimum(x):
    #2*x+20 =0, x = -20/2--> -10
    return -10

x = np.linspace(-50,50,100)
y = func(x)


#Where is the minimum
#x0 = np.random.choice(x,1)
x0 = np.random.randint(-50,-40)  #Nagative
#x0 = np.random.randint(40,50) #Postive
iterations = 100
lr = 0.1



points_gr = []

fig, ax = plt.subplots(figsize=(10, 4))
ax.set_title("Example x^2 function: Finding Minimum using Gradient Descent")
ax.plot(x, y)
ax.scatter(x0, func(x0), color="red", label='Starting Point')
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
x1 = x0
for ind in range(iterations):
    dx_val = derivative(x1)
    x1 = x1 - lr*dx_val
    ax.scatter(x1, func(x1), color="blue", label='Minimum Point')
    points_gr.append(x1)



print("minimum:",x1)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
x0 = np.random.randint(40,50) #Postive
ax.set_title("Example x^2 function: Finding Minimum using Minimum")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.plot(x, y)
ax.scatter(x0, func(x0), color="red", label='Starting Point')
xmin = get_minimum(x0)
ax.scatter(xmin, func(xmin), color="green", label='Starting Point')
plt.show()
```
