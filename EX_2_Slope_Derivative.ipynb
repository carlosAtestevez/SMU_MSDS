```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def function_x(x):
    return 2*x**2

def derivative_x(x):
    return 4*x

def calculate_tangent_line(m,x1,x2,y1):
    y2 = m*(x2-x1)+y1
    return y2

#Drawing initial plot
x = np.linspace(1,100,100)
y = function_x(x)
fig,ax = plt.subplots(figsize = (10,4))
ax.set_title("Tangent and Derivative")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.plot(x,y,label='y = x^2')


#Calculating slope
x_1_list = [20,60,100,120]
for x_1 in x_1_list:
    #x_1 = 60
    y_1 = function_x(x_1)
    m = derivative_x(x_1)
    print(f"X = {x_1}, Slope = {m}")
    x_2 = np.linspace(x_1-15,x_1+15,50)
    tangent_line = calculate_tangent_line(m,x_1,x_2,y_1)
    plt.plot(x_2,tangent_line,color="green",linestyle='--')
    ax.scatter(x_1,y_1,color="red")

plt.legend()
plt.show()
```
