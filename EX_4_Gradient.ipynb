```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def func(x, y):
    return 100 + x**2 + y**2 + 20*x + 20*y

# Define the gradient of the function
def gradient(x, y):
    df_dx = 2*x + 20
    df_dy = 2*y + 20
    return np.array([df_dx, df_dy])

# Generate data for the surface plot
x = np.linspace(-50, 50, 100)
y = np.linspace(-50, 50, 100)
x, y = np.meshgrid(x, y)
z = func(x, y)

# Initialize gradient descent parameters
#x0, y0 = np.random.choice(np.linspace(-50, 50, 100), 1)[0], np.random.choice(np.linspace(-50, 50, 100), 1)[0]  # Starting point
x0 = np.random.randint(-50,-40)
y0 = np.random.randint(40,50)

lr = 0.01  # Learning rate
iterations = 100  # Number of iterations

# Arrays to store the points for plotting
x_vals = [x0]
y_vals = [y0]
z_vals = [func(x0, y0)]

# Perform gradient descent
for i in range(iterations):
    grad = gradient(x0, y0)
    x0 = x0 - lr * grad[0]
    y0 = y0 - lr * grad[1]
    
    # Store the updated points
    x_vals.append(x0)
    y_vals.append(y0)
    z_vals.append(func(x0, y0))


# Convert lists to arrays for plotting
x_vals = np.array(x_vals)
y_vals = np.array(y_vals)
z_vals = np.array(z_vals)

# Plotting the 3D surface
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.6)
ax.set_title("Gradient Descent Path")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z(x,y)')

# Plotting the gradient descent path
ax.plot(x_vals, y_vals, z_vals, color='red', marker='o', label='Gradient Descent Path')
ax.scatter(x_vals, y_vals, z_vals, color='blue')

plt.legend()
plt.show()
```
